# Analysis of Repressions Against Cultural Figures

This notebook analyzes a dataset of **cultural workers who were sentenced to imprisonment** in prisons or penal colonies.

The goal is to clean the dataset, extract the year of each sentence, count how many people were sentenced in each year, and visualize the results.

### 1. Loading the data

The notebook begins by importing pandas and loading the dataset from repressions_culture.csv.

```python
import pandas as pd
```

```python
Cultural figures sentenced to serve their sentences in prisons and penal colonies
df = pd.read_csv('repressions_culture.csv')
```

```python
df
```

### 2. Cleaning the dataset

The first row of the CSV contains the actual column names, so I use that row as the header and remove it from the data.

This makes the dataframe easier to work with because each column now has a meaningful name.

```python
# Use first row as column names
df.columns = df.iloc[0]

# Remove the row that became the header
df = df.iloc[1:].reset_index(drop=True)
```

```python
df
```

### 3. Extracting sentencing years

The column Прысуд contains text descriptions of each sentence.
I use a regular expression to extract the first four-digit year from each sentence and save it in a new column called Год прысуду.

```python
df['Прысуд']
```

```python
# Extract first 4-digit year from the sentence
df['Год прысуду'] = df['Прысуд'].str.extract(r'(\d{4})')
```

```python
df['Год прысуду']
```

### 4. Counting sentences by year

After extracting the years, I count **how many people were sentenced** in each year.

The result is a new dataframe called **year_counts**, with two columns:

- year: the sentencing year
- count: the number of people sentenced that year

```python
year_counts = (
    df['Год прысуду']
    .value_counts()
    .sort_index()
    .reset_index()
)

year_counts.columns = ['year', 'count']
```

### 5. Visualizing the results

Finally, I use Altair to create a bar chart showing the number of sentenced cultural figures by year.

The chart makes it easier to see how sentencing patterns changed over time.

```python
import altair as alt

chart = (
    alt.Chart(year_counts)
    .mark_bar()
    .encode(
        x=alt.X('year:O', title='Year'),
        y=alt.Y('count:Q', title='Number of people'),
        tooltip=['year', 'count']
    )
    .properties(
        title='People sentenced by year',
        width=600,
        height=400
    )
)

chart
```
